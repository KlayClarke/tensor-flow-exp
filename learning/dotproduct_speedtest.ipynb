```python
import numpy as np
from datetime import datetime

a = np.random.randn(100)
b = np.random.randn(100)
T = 1000000

print(a)
print(b)
```

```python
def slow_dot_product(a,b):
  result = 0
  for e, f in zip(a, b):
    result += e*f
  return result
```

```python

# PYTHON TEST

# record initial time
t0 = datetime.now()
# run test
for t in range(T):
  slow_dot_product(a, b)
# calculate time elapsed
dt1 = datetime.now() - t0 
```

```python
# NUMPY TEST

# record initial time
t0 = datetime.now()
# run test
for t in range(T):
  a.dot(b)
# calculate time elapsed
dt2 = datetime.now() - t0
```

```python
# calculate time differential
print("dt1/dt2:", dt1.total_seconds()/dt2.total_seconds())
```
