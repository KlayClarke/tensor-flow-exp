```python
import numpy as np
import matplotlib.pyplot as plt
```

## Line Chart

```python
# create one dimensional array with 1000 evenly spaced pointer between 0 and 20

x = np.linspace(0,20,1000)
```

```python
# generate fake data for Y Axis (this case would gen a sine wav)

y = np.sin(x) + .2 * x
```

```python
# plot data

plt.plot(x,y)
```

```python
# label axes

plt.plot(x,y)
plt.xlabel("input")
plt.ylabel("output")
plt.title("random plot");
```

## Scatterplot

```python
# 100 data points with dimensionality 2

x = np.random.randn(100,2)
```

```python
# to create scatter plot

plt.scatter(x[:,0], x[:,1])
```

```python
# create scatter plot with 200 data points

x = np.random.randn(200,2)
```

```python
# center first half of data points at 3,3

x[:50] += 3
```

```python
# create array of zeros the size of 200

y = np.zeros(200)
```

```python
# set first half of these to 1

y[:50] = 1
```

```python
# c variable allows you to color the plot

plt.scatter(x[:,0],x[:,1], c=y)
```
